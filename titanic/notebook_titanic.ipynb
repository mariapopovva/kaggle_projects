```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

```python
data
```

```python
data['Initial']=0
for i in data:
    data['Initial']=data.Name.str.extract('([A-Za-z]+)\.')

data['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)

data.loc[(data.Age.isnull())&(data.Initial=='Mr'),'Age']=33
data.loc[(data.Age.isnull())&(data.Initial=='Mrs'),'Age']=36
data.loc[(data.Age.isnull())&(data.Initial=='Master'),'Age']=5
data.loc[(data.Age.isnull())&(data.Initial=='Miss'),'Age']=22
data.loc[(data.Age.isnull())&(data.Initial=='Other'),'Age']=46

data['Embarked'].fillna('S',inplace=True)

data['Age_band']=0
data.loc[data['Age']<=16,'Age_band']=0
data.loc[(data['Age']>16)&(data['Age']<=32),'Age_band']=1
data.loc[(data['Age']>32)&(data['Age']<=48),'Age_band']=2
data.loc[(data['Age']>48)&(data['Age']<=64),'Age_band']=3
data.loc[data['Age']>64,'Age_band']=4

data.head()
data['Family_Size']=0
data['Family_Size']=data['Parch']+data['SibSp']
data['Alone']=0
data.loc[data.Family_Size==0,'Alone']=1
data['Fare_Range']=pd.qcut(data['Fare'],4)
data['Fare_band']=0
data.loc[data['Fare']<=7.91,'Fare_band']=0
data.loc[(data['Fare']>7.91)&(data['Fare']<=14.454),'Fare_band']=1
data.loc[(data['Fare']>14.454)&(data['Fare']<=31),'Fare_band']=2
data.loc[(data['Fare']>31)&(data['Fare']<=513),'Fare_band']=3
data['Sex'].replace(['male','female'],[0,1],inplace=True)
data['Embarked'].replace(['S','C','Q'],[0,1,2],inplace=True)
data['Initial'].replace(["Mr","Mrs","Miss","Master", "Other","Dona"],[0,1,2,3,4,5],inplace=True)
data.drop(['Name','Age','Ticket','Fare','Cabin','Fare_Range','PassengerId'],axis=1,inplace=True)
```

```python
data
```

```python
test2 = test.copy()

test2["Initial"] = test2["Name"].str.extract("([A-Za-z]+)\.")
test2["Initial"].replace(
    ['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],
    ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],
    inplace=True
)

test2.loc[(test2.Age.isnull())&(test2.Initial=='Mr'),'Age']=33
test2.loc[(test2.Age.isnull())&(test2.Initial=='Mrs'),'Age']=36
test2.loc[(test2.Age.isnull())&(test2.Initial=='Master'),'Age']=5
test2.loc[(test2.Age.isnull())&(test2.Initial=='Miss'),'Age']=22
test2.loc[(test2.Age.isnull())&(test2.Initial=='Other'),'Age']=46

test2["Embarked"].fillna("S", inplace=True)
test2["Fare"].fillna(test2["Fare"].median(), inplace=True)

test2["Age_band"]=0
test2.loc[test2["Age"]<=16,"Age_band"]=0
test2.loc[(test2["Age"]>16)&(test2["Age"]<=32),"Age_band"]=1
test2.loc[(test2["Age"]>32)&(test2["Age"]<=48),"Age_band"]=2
test2.loc[(test2["Age"]>48)&(test2["Age"]<=64),"Age_band"]=3
test2.loc[test2["Age"]>64,"Age_band"]=4

test2["Family_Size"] = test2["Parch"] + test2["SibSp"]
test2["Alone"] = (test2["Family_Size"]==0).astype(int)

test2["Fare_band"]=0
test2.loc[test2["Fare"]<=7.91,"Fare_band"]=0
test2.loc[(test2["Fare"]>7.91)&(test2["Fare"]<=14.454),"Fare_band"]=1
test2.loc[(test2["Fare"]>14.454)&(test2["Fare"]<=31),"Fare_band"]=2
test2.loc[test2["Fare"]>31,"Fare_band"]=3

test2["Sex"].replace(["male","female"],[0,1], inplace=True)
test2["Embarked"].replace(["S","C","Q"],[0,1,2], inplace=True)
test2["Initial"].replace(["Mr","Mrs","Miss","Master", "Other","Dona"],[0,1,2,3,4,5], inplace=True)

test2.drop(["Name","Ticket","Cabin","Age","Fare"], axis=1, inplace=True)
```

```python
y = data["Survived"]
X = data.drop(columns=["Survived"])
test_ids = test["PassengerId"].copy()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
y = data["Survived"]
X = data.drop(columns=["Survived"])
X_test = test2.drop(columns=["PassengerId"])

model = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
model.fit(X, y)
pred = model.predict(X_test)

submission = pd.DataFrame({
    "PassengerId": test_ids,
    "Survived": pred.astype(int)
})
submission.to_csv("submission_dt_depth3_v2.csv", index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42, verbose=2, max_depth=3, max_features=2)
model.fit(X, y)
pred = model.predict(X_test)

submission = pd.DataFrame({
    "PassengerId": test_ids,
    "Survived": pred.astype(int)
})
submission.to_csv("submission_RF.csv", index=False)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y, cv=skf, scoring="accuracy")
f1 = cross_val_score(model, X, y, cv=skf, scoring="f1")

print("F1 mean:", f1.mean())
print("Accuracy по фолдам:", scores)

print("CV:", round(scores.mean(),4))
print("CV std:", round(scores.std(),4))
```
